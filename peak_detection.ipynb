```python
from scipy.signal import find_peaks
```

```python
@contextlib.contextmanager
def plot(title):
    plt.figure(figsize=(20,5))
    plt.title(title)
    yield
    plt.xlim(0, 8192)
    ylim = 40
    plt.ylim(0,ylim)
    plt.grid(True)
    plt.show()
    
with plot('scipy find_peak'):
    plt.plot(evlp)
    
    peaks, _ = find_peaks(
        evlp, 
        distance=100, 
        prominence=10
    )
    plt.scatter(peaks, evlp[peaks], color='orange')
```
