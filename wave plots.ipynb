```python
from glob import glob

waves = []
for i, fname in enumerate(glob("*.csv")):
    plt.figure(figsize=(15,3))
    d = np.loadtxt(fname)
    
    waves.append(d)
    plt.plot(d)
    plt.ylabel(i)
    plt.show()
```

```python
def plot(wave):
    plt.figure(figsize=(15,3))
    plt.plot(wave)
#     plt.ylabel(i)
    plt.show()
```

```python
d1 = waves[10]
d2 = waves[11]

plot(d1)
plot(d2)
```

```python
def qiaodu(d):
    rms = np.sqrt(np.mean(d**2))
    ave = np.mean(d)
    denom = rms ** 4
    numer = np.mean((np.abs(d)-ave)**4)
#     print(numer)
#     print(denom)
    return numer/denom

print(qiaodu(d1))
print(qiaodu(d2))
```

```python
for i, d in enumerate(waves):
    if i in [0,8]:
        continue
    plt.figure(figsize=(15,3))
    plt.grid(True)
    plt.ylim(-4,4)
    plt.plot(d)
    plt.text(0,2.5,f'index={i}; $C_q$={qiaodu(d):.2f}', size=25)
#     plt.ylabel(f'{i} - {qiaodu(d):.2f}')
    plt.show()
```
